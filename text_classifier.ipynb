## This file explains how to do natural language process



```python
import torch
import torch.nn as nn
import torch.optim as optim

# -----------------------------
# 1. Tiny dataset
# -----------------------------

texts = [
    "i love this movie",
    "this movie is good",
    "i really like this",
    "amazing film",
    "i hate this movie",
    "this movie is bad",
    "i really dislike this",
    "terrible film"
]

labels = [
    1,  # positive
    1,
    1,
    1,
    0,  # negative
    0,
    0,
    0
]

# -----------------------------
# 2. Tokenization
# -----------------------------

def tokenize(text):
    return text.lower().split()

tokenized_texts = []

for text in texts:
    tokens = tokenize(text)
    tokenized_texts.append(tokens)

print(tokenized_texts)
```

# create a word to number dictionary with each word a unique ID number

## build vocabulary

```python
# -----------------------------
# 3. Build vocabulary
# -----------------------------

# We will create a vocabulary that maps each unique token to a unique integer index.
# We will also include special tokens for padding ("<PAD>") and unknown words ("<UNK>").
# The "<PAD>" token will be used to pad sequences to a fixed length, 
# and the "<UNK>" token will be used for any tokens that are not in the vocabulary.
vocab = {
    "<PAD>": 0,
    "<UNK>": 1
}

# Build the vocabulary from the tokenized texts
# We will iterate through each token in the tokenized texts and add it to the vocabulary if it is not already present.
for tokens in tokenized_texts:
    for token in tokens:
        if token not in vocab:
            vocab[token] = len(vocab)

print(vocab)
```

```python
# -----------------------------
# 4. Integer encoding
# -----------------------------
# 4. Integer encoding
# -----------------------------
# Now that we have a vocabulary, we can convert each token in the tokenized texts to its corresponding integer index.
# We will create a function called `encode` that takes a list of tokens and
# the vocabulary as input and returns a list of integer indices.
def encode(tokens, vocab):
    ids = []
    for token in tokens:
        # We will use the `get` method of the dictionary to retrieve the index of the token from the vocabulary.
        # If the token is not found in the vocabulary, we will return the index of the "<UNK>" token.
        # This way, we can handle any tokens that are not in the vocabulary without causing an error.
        ids.append(vocab.get(token, vocab["<UNK>"]))
    return ids

#encoded_texts = [encode(tokens, vocab) for tokens in tokenized_texts]

encoded_texts = []
for tokens in tokenized_texts:
    ids = encode(tokens, vocab)
    encoded_texts.append(ids)

print(encoded_texts)
```

```python
# -----------------------------
# 5. Padding
# -----------------------------
# Since the encoded texts have different lengths, 
# we need to pad them to a fixed length before we can feed them into a neural network.
# We will pad the sequences with the index of the "<PAD>" token (which is 0) to make them all the same length.

max_length = max(len(ids) for ids in encoded_texts)

# We will create a function called `pad` that takes a list of integer indices 
# and the maximum length as input and returns a padded list of integer indices.
def pad(ids, max_length):
    padding_needed = max_length - len(ids)
    # We will add the necessary number of "<PAD>" tokens 
    # (which have an index of 0) to the end of the sequence to make it the same length as `max_length`.
    return ids + [0] * padding_needed

# We will apply the `pad` function to each of the encoded texts to get a list of padded integer sequences.
padded_texts = [pad(ids, max_length) for ids in encoded_texts]

print(padded_texts)
```
